# Parte 1 - Construir a CNN

```
# Importar pacotes
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```
# Inicializar a CNN
classifier = Sequential()

# Passo 1 - Convolução
classifier.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))

# Passo 2 - Max Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Segunda camada de convolução e max pooling
classifier.add(Conv2D(32, (3, 3), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Passo 3 - Flattening
classifier.add(Flatten())

# Passo 4 - Full connection
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'sigmoid'))
```

```
# Compilar a CNN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
# Visualizar o modelo compilado
classifier.summary()
```

# Parte 2 - Treinamento

```
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```
# Base de treino
training_set = train_datagen.flow_from_directory('D:\Manoela\Dropbox\Aulas\Deep Learning/exercicio/CNN/dataset_pipoca/training_set',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

```
# Base de teste
test_set = test_datagen.flow_from_directory('D:\Manoela\Dropbox\Aulas\Deep Learning/exercicio/CNN/dataset_pipoca/test_set',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

```
#Treinar

classifier.fit_generator(training_set,
                         steps_per_epoch = 26,
                         epochs = 25,
                         validation_data = test_set,
                         validation_steps = 6)
```

# Parte 3 - Fazendo novas previsões
