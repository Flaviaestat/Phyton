```python
def extrai_lat_long(cep):
    cep_str = str(cep)
    endereco = pycep_correios.get_address_from_cep(cep_str)
    geolocator = Nominatim(user_agent="myGeocoder")
    location = geolocator.geocode(endereco['logradouro'] + ", " + endereco['cidade'] + " - " + endereco['bairro'])
    return(location.latitude, location.longitude)
```

```python
ceps = pd.read_csv('zip_codes.csv')

#colocar zeros a esquerda
```

```python
ceps.head()
```

```python
lat = []
long = []

ceps['nr_zip_code'] = ceps['nr_zip_code'].astype('string').str.zfill(8)

for i in range(0, len(ceps)):
#     print(i)
    cep = ceps['nr_zip_code'][i]
    try:
        latitude, longitude = extrai_lat_long(cep)
    except:
        latitude = 0
        longitude = 0
    lat.append(latitude)
    long.append(longitude)
    if i/5 == 0:
        time.sleep(5)
    else:
        pass

ceps['lat'] = lat
ceps['long'] = long
```

```python
import pycep_correios
from geopy.geocoders import Nominatim
import numpy as np
import pandas as pd
import time
from geopy.extra.rate_limiter import RateLimiter
```

```python
geolocator = Nominatim(user_agent="test_app")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds = 1)
```

```python
import acessos as ac

query = """
        select distinct username as cpf, customer_id,
        upper(state_abbreviation) as estado,
        upper(locality) as cidade,
        upper(public_place) ||' , ' || upper(locality) || ' - ' || upper(state_abbreviation) as endereco	
        from "processed-zone-database-customer".customer_address a
		left join "processed-zone-database-access-control".user u on (u.id = customer_id)
        where UPPER(locality) = 'SALVADOR' and state_abbreviation = 'BA'

        """
df = ac.df_athena('flavia-costa', query)
```

```python
df['geocode'] = df['endereco'].apply(geocode)
```

```python
df['latitude'] = df['geocode'].apply(lambda loc: loc.latitude if loc else None)
df['longitude'] = df['geocode'].apply(lambda loc: loc.longitude if loc else None)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df2 = df.query('latitude <0')
```

```python
len(df2)
```

```python
#rodar K-means
from sklearn.cluster import KMeans
```

```python
#from sklearn.neighbors import NearestNeighbors
kmeans = KMeans(n_clusters=8).fit(df2[['latitude', 'longitude']])
cluster = kmeans.predict(df2[['latitude', 'longitude']])
```

```python
#plotar gráfico
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize=(20,20))
plt.subplot(223)
plt.scatter(df2['latitude'], df2['longitude'], c=cluster)
plt.legend()
```

```python
df2['cluster'] = cluster
```
