```python
query_grafico_diario = ("""
                         
			CREATE TABLE public.daily_spending_pag_will AS 
			with datasref as (			
			select distinct			
			substring(cast(dt_autorizacao as varchar), 1, 10) as dia,			
			substring(cast(dt_autorizacao as varchar), 1, 7) as mes			
			from platform_curated_zone.authorization_pag			
			where to_date(substring(cast(dt_autorizacao as varchar), 1, 10), 'yyyy-mm-dd') >= to_date('2021-01-01', 'yyyy-mm-dd')		
			)			
			, uso_credito_pag as (			
			--gera spending pag			
			select			
			nr_cpf,			
			substring(cast(dt_autorizacao as varchar), 1, 10) as dia,			
			sum(vl_real) as spending			
			from	platform_curated_zone.authorization_pag		
			where	is_saque = false	
                    and	is_cancelada = false		
                    and	is_recarga = false		
                    and	is_pag_limite_utilizado = 'C'
                    and ds_customer = 'pag'
			group by 1, 2			
			)			
			, uso_credito_will as (			
			--gera spending will			
			select	nr_cpf,		
			substring(cast(dt_autorizacao as varchar), 1, 10) as dia,			
			sum(vl_real) as spending			
			from	platform_curated_zone.authorization_will		
			where   ds_transacao = 'credito'			
			and	ds_status_compra = 'aprovada'		
			and	cd_retorno = '00'
			and ds_customer = 'will'
            and ds_mcc != 'TRANSACOES WEBSERVICE'
			group by 1, 2			
			)			
			--append das duas			
			, uso_credito_geral as (
			select  *			
			from  uso_credito_pag			
			union all			
			select  *			
			from  uso_credito_will			
			)			
			select			
			d.mes,			
			substring(cast(d.dia as varchar), 9, 2) as dia_mes,			
			count(distinct nr_cpf) as ativos,			
			sum(spending) as spending			
			from datasref d			
			left join uso_credito_geral c on (c.dia <= d.dia and substring(cast(c.dia as varchar), 1, 7) = d.mes)			
			group by 1,2
			order by 1,2
                         
                         """)
```
